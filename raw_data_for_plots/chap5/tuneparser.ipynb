```python
import os
import sys
import numpy as np
import matplotlib as mpl
import matplotlib.pyplot as plt
import pandas as pd
```

```python
def parse(filename):
    tps = []
    lat = []
    current_tps = -1
    current_lat = -1
    with open(filename,'r') as f:
        line = f.readline()
        while line:
            if line.startswith('current_performance(tps, lat)'):
                meta = line.split(' ')
                tps_temp = float(meta[2])
                lat_temp = float(meta[3])
                if(current_tps < 0 or current_tps < tps_temp):
                    current_tps = tps_temp
                if(current_lat < 0 or current_lat > lat_temp):
                    current_lat = lat_temp
                tps.append(current_tps)
                lat.append(current_lat)
            line = f.readline()
    return tps, lat
```

```python
knobs_14_tps, knobs_14_lat = parse('tunning_with14knbos.log')
plt.plot(knobs_14_tps)
knobs_42_tps, knobs_42_lat = parse('tunning_with42knbos.log')
plt.plot(knobs_42_tps)
plt.show()
```
